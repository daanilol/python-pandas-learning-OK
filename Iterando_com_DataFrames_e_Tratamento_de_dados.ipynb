# Iterando com DataFrames

```python
import pandas as pd
```

```python
dataset = pd.read_csv('db.csv', sep=';', index_col = 0)
```

```python
dataset.head()
```

```python
for item in dataset:
  print(item)
```

```python
dataset.iterrows()
```

```python
for index, row in dataset.iterrows():
  if(2019 - row['Ano'] != 0):
    dataset.loc[index, 'km_media'] = row['Quilometragem'] / (2019 - row['Ano'])
  else:
    dataset.loc['index', 'km_media'] =  0 

dataset
```
