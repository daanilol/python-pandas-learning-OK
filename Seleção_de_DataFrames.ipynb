# Selecionando linhas - [i:j]

**Observação:** A indexação tem origem no zero e nos fatiamentos (slices) a linha com índice i é **incluída** e a linha com índice j **não é incluida** no resultado.

```python
dataset[:3]
```

# Utilizando o *.loc* para seleções

```python
dataset
```

```python
dataset.loc['Passat']
```

```python
#filtragem [[index_1, index_2], [column_1, column_2]]
dataset.loc[['Passat', 'DS5'], ['Motor', 'Valor']]
```

```python
#filtragem de dados [index, [column_1, column_2]]
dataset.loc[:, ['Motor', 'Valor']]
```

# Utilizando *.iloc* para seleções

**Observação:** Seleciona com base nos índices, ou seja, se baseia na posição das informações.

Localização baseada em indexação númerica.

```python
dataset.head()
```

```python
dataset.iloc[1:4]
```

```python
dataset.iloc[1:4, [0, 5, 2]]
```

```python
dataset.iloc[[1, 42, 22], [0, 5, 2]]
```
